```python
import pandas as pd
df =pd.read_csv('Turnoutvotes.csv')
df
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.plot('Year','female',data=df, marker='', color='olive', linewidth=2,label='Female Voters Turnout %')
plt.plot('Year','male',data=df, marker='', color='red', linewidth=2,label='Male Voters Turnout %')
plt.plot('Year','total',data=df,marker='o', markerfacecolor='blue', markersize=8, color='skyblue', linewidth=2,label='Total Voters Turnout %')
plt.xlabel("Years")
plt.ylabel("Percentage")
plt.legend()
plt.show
```

```python
plt.plot('LitYear','Litfemale',data=df, marker='', color='red', linewidth=2,label='Female litracy %')
plt.plot('LitYear','Litmale',data=df, marker='', color='blue', linewidth=2,label='Male litracy %')
plt.plot('Year','female',data=df, marker='o', color='red', linewidth=2)
plt.plot('Year','male',data=df, marker='o', color='blue', linewidth=2)
plt.xlabel("Years")
plt.ylabel("Percentage")
plt.legend()
plt.show
```

```python
plt.plot('LitYear','Littotal',data=df, marker='', color='brown', linewidth=3,label='Litracy Rate %')
plt.plot('Year','total',data=df,marker='', color='skyblue', linewidth=3,label='Turnout-Vote %')
plt.xlabel("Years")
plt.ylabel("Percentage")
plt.legend()
plt.show
```

```python
df['diff'] = df['male'] - df['female']
df['Litdiff'] = df['Litmale'] - df['Litfemale']
plt.plot('Year','diff',data=df,marker='o', color='green', linewidth=2,label='Turnout-Vote Gender-Gap %')
plt.plot('LitYear','Litdiff',data=df,marker='o', color='blue', linewidth=2,label='Litracy Gender-Gap %')
plt.xlabel("Years")
plt.ylabel("Percentage")
plt.legend()
plt.show
```
